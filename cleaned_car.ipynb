```python
car['price'] = car['price'].astype(str)                      # Convert to string
car['price'] = car['price'].str.replace('$', '', regex=False)
car['price'] = car['price'].str.replace(',', '', regex=False)
car['price'] = car['price'].astype(int)                      # Convert to int
car['price']
```

```python
car.info()
```

```python
car['milage'] = car['milage'].astype(str)  # Ensure it's string
car['milage'] = car['milage'].str.replace(',', '', regex=False)  # Remove commas
car['milage'] = car['milage'].str.replace(' mi.', '', regex=False)  # Remove " mi."
car['milage'] = car['milage'].astype(int)  # Convert to integer

car['milage']
```

```python
car = car[car['fuel_type'].isin(['E85 Flex Fuel', 'Gasoline', 'Hybrid', 'Diesel', 'Plug-In Hybrid'])]
car
```

```python
car.drop(columns=['clean_title'], inplace=True)  # Drop the 'clean_title' column
```

```python
car['accident']=car['accident'].isin(['At least 1 accident or damage reported', 'None reported'])
```
