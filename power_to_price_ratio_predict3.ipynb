```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
file_path = 'processed_used_cars.csv'
df=pd.read_csv(file_path)
```

```python
df.head()
```

```python
# create more features
df["price_per_mile"] = df["price_numeric $"] / (df["milage_numeric"]+1)
df["price_per_year"] = df["price_numeric $"] / df["vehicle_age"]
df['mile_per_year'] = df["milage_numeric"] / df["vehicle_age"]
```

```python
df.head()
```

```python
features=['price_numeric $','milage_numeric','engine_hp HP','model_year','vehicle_age','price_per_mile','price_per_year','mile_per_year']
x=df[features]
y=df['power_to_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))
```

```python
x1 = df[["engine_hp HP"]].values.reshape(-1, 1)
x2 = df[["engine_hp HP", "mile_per_year"]].values.reshape(-1, 2)
y = df["power_to_price"].values.reshape(-1, 1)
```

```python
x1_train, x1_test, x2_train, x2_test, y_train, y_test = train_test_split(x1, x2, y, random_state=42)
```

```python
model1 = LinearRegression()
model2 = LinearRegression()
```

```python
model1.fit(x1_train, y_train)
model2.fit(x2_train, y_train)
```

```python
predict1 = model1.predict(x1_test)
predict2 = model2.predict(x2_test)
```

```python
# Score the predictions with mse and r2
mse1 = mean_squared_error(y_test, predict1)
r21 = r2_score(y_test, predict1)
mse2 = mean_squared_error(y_test, predict2)
r22 = r2_score(y_test, predict2)
print(f"Single Feature:")
print(f"mean squared error (MSE): {mse1}")
print(f"R-squared (R2): {r21}")
print("---------------------")
print(f"Two Features:")
print(f"mean squared error (MSE): {mse2}")
print(f"R-squared (R2): {r22}")
print("---------------------")
print(f"Difference: {r21-r22}")
```
